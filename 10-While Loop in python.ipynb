# 10-While Loop in python

While loop is a statement that will execute its block of code as long as its condition remains true.

### Infinite loop

```python
i=1
while i==1:
    print("It's an infinite loop")   
```

### To get an user input other than null

```python
name = ""

while len(name) == 0:
  name = input("enter your name: ")
print("hello "+name)
```

### Aliter method

```python
name = None
while not name:
  name = input("enter your name: ")
print("hello "+name)
```

example to print 1 to 100 using while loop

```python
i=1
while i<=100:
    print(i)
    i+=1
```

Here we added i += 1 so that the value of i increases by 1 for every loop execution.If we don't add i+=1 then the loop will run for infinite times printing the initial i value which is 1.
